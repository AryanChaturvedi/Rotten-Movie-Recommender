## THE 84 BEST SUPERHERO MOVIES OF ALL TIME

```python
#load packages
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL of the site
base = 'https://editorial.rottentomatoes.com/guide/best-superhero-movies-of-all-time/'
```

```python
# sending a request to the webpage
r = requests.get(base)
r.status_code
```

```python
html = r.content
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
with open('THE 84 BEST SUPERHERO_LXML_Parser.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
# Find all div tags on the webpage containing the information we want to scrape
divs = soup.find_all("div", {"class": "col-sm-18 col-full-xs countdown-item-content"})
divs
```

```python
divs[0].find("h2")
```

```python
headings = [div.find("h2") for div in divs]
```

```python
movie_names = [heading.find('a').string for heading in headings]
movie_names
```

```python
years = [heading.find("span", class_ = 'start-year').string for heading in headings]
years = ["2000" if x == '()' else x for x in years]
years
```

```python
# removing brackets from year
years = [year.strip('()') for year in years]
#years = [int(year) for year in years]
years
```

```python
# finding tomatometer score from site
#score is located inside heading of span element with class tMeterScore
#if this heading is absent in divs then fill score=69
scores = [heading.find("span", class_ = 'tMeterScore').string for heading in headings]
scores = ["69" if x is None else x for x in scores]
scores = [s.strip('%') for s in scores]  #removing %sign from score


#scores = [int(s) for s in scores]
scores
```

```python
#if consensus div is unavailable then add empty string else get div element that data
consensus = [soup.new_string("") if div.find("div", {"class": "info critics-consensus"}) is None else div.find("div", {"class": "info critics-consensus"}) for div in divs]
consensus
```

```python
#scrapping texts from consensus div
[con.text for con in consensus]
```

```python
common_phrase = 'Critics Consensus: '
len(common_phrase)
```

```python
consensus[0].text
```

```python
#removing Critics Consensus: from each texts
consensus[0].text[19:]
```

```python
common_len = len(common_phrase)
consensus_text = [con.text[common_len:] for con in consensus]
consensus_text
```

```python
# We can add if-else logic to only truncate the string in case it starts with the common phrase
consensus_text = [con.text[common_len:] if con.text.startswith(common_phrase) else con.text for con in consensus ]
consensus_text
```

```python
divs[0]
```

```python
div_casts= [div.find("div",{"class": "info cast"}) for div in  divs]
casts_link = div_casts[0].find_all("a")
cast_names = [link.string for link in casts_link]
```

```python
# cast = []
# for c in div_casts:
#     casts_link = c.find_all("a")
#     cast_names = [link.string for link in casts_link]
#     cast.append(", ".join(cast_names)) # Joining is optional
    
# cast
#cast_info = [div.find("div", class_ = 'cast') for div in divs]
```

```python
cast =  [soup.new_string("") if c is None else  ", ".join([link.string for link in c.find_all("a")]) for c in div_casts]
cast
```

```python
# Extracting all director divs
directors= [soup.new_tag("div") if div.find("div",class_= "info director") is None else div.find("div",class_= "info director") for div in divs]
directors
```

```python
director= []

for d in directors:
    if d!=None:
        director_link = d.find_all("a")
        director_name = [link.string for link in director_link]
        director.append(", ".join(director_name)) # Joining is optional
    
    
    
director
```

```python
tag_div = soup.find("div")
count_children = len(tag_div.find_all())
print(count_children)
```

```python
synopsis = [soup.new_string("") if div.find("div",class_="info synopsis") is None else  div.find("div",class_="info synopsis") for div in divs]
synopsis[0]
```

```python
cm_phrase = "Synopsis:"
cm_len=len(cm_phrase)
```

```python
synopsis_texts = [ syn.text[cm_len:] for syn in synopsis]
synopsis_texts
```

```python
genre=[]
for i in range(84):
    genre.append("Super-hero")
```

```python
Platform=[]
for i in range(84):
    
        Platform.append("public")
    
```

```python
# load the pandas package
import pandas as pd
```

```python
# A dataframe is a tabular data type, frequently used in data science

movies_info = pd.DataFrame()
movies_info  # The dataframe is still empty, we need to fill it with the info we gathered
```

```python
movies_info["Movie Title"] = movie_names
movies_info["Year"] = years
movies_info["Score"] = scores
movies_info["Director"] = director
movies_info["Synopsis"] = synopsis_texts    # Homework
movies_info["Cast"] = cast
movies_info["Genre"] = genre
movies_info["Critic_Consensus"] = consensus_text
movies_info['Platform']=Platform
# Let's see how it looks
movies_info
```

```python
movies_r = pd.read_csv("movies_r.csv")
movies_r
```

```python
movies_r = pd.concat([movies_r, movies_info], ignore_index = True)
movies_r.reset_index()
movies_r
```

```python
#  write data to CSV file
movies_r.to_csv("movies_r.csv", index = False, header = True)
```
